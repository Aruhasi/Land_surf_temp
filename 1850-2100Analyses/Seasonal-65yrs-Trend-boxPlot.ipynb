```python
# grouped boxplot
import seaborn as sns

# create a list of the three categories
categories = ['DJF', 'MAM', 'JJA', 'SON']

# create a list of the three datasets
data1= [DJF1_65yr,MAM1_65yr,JJA1_65yr, SON1_65yr]
data = [DJF_65yr,MAM_65yr,JJA_65yr, SON_65yr]

data_scatter1_CRU ={'DJF':[0.33],
                'MAM':[0.36],
                'JJA':[0.40],
                'SON':[0.64]}
data_scatter1_MLOST ={'DJF':[0.91],
                'MAM':[0.69],
                'JJA':[0.46],
                'SON':[0.86]}

data_scatter_CRU = {'DJF':[2.09],
                'MAM':[2.09],
                'JJA':[1.68],
                'SON':[1.61]}

data_scatter_MLOST = {'DJF':[2.17],
                'MAM':[2.33],
                'JJA':[1.71],
                'SON':[2.02]}
# create a list of the colors for each category
colors = ['red', 'green', 'blue','m']
colors1 =['#580F41','#4B0082']
markers = ['o','X']
# create a figure with two subplots
fig, ax1 = plt.subplots(ncols=1, figsize=(10, 8))


sns.boxplot(data=data1, palette="Set3", ax=ax1,width=0.5)
    # overlay a stripplot for the current subplot

sns.stripplot(data=data_scatter1_CRU, color=colors1[0], jitter=True, marker=markers[0], size=7, alpha=0.7)
sns.stripplot(data=data_scatter1_MLOST, color=colors1[1],jitter=True, marker=markers[1], size=7, alpha=0.7)


    # set the x-axis label for the current subplot
ax1.set_xlabel('Category')
ax1.set_xticklabels(['DJF', 'MAM', 'JJA', 'SON'])
    # set the y-axis label for the current subplot
ax1.set_ylabel('Temperature Trend (°C/65yr)')
    # set the title for the current subplot
ax1.set_title(f'1892-1957')

# show the plot
plt.show()
```
