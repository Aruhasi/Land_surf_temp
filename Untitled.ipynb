```python
%matplotlib inline
import os
import numpy as np
import xarray as xr
import pandas as pd
import scipy.signal
# from cdo import *
# cdo = Cdo()

from metpy.units import units
from scipy import stats
from scipy.stats import linregress

# import cdstoolbox as ct

data_GHCNv4   = '/work/mh0033/m301036/LSAT/Data/gistemp250_GHCNv4.nc'
data_CRUTEMP  = '/work/mh0033/m301036/LSAT/Data/CRUTEM.5.0.1.0.anomalies.nc'
data_MLOST    = '/work/mh0033/m301036/LSAT/Data/air.mon.anom.nc'
data_Berkely  = '/work/mh0033/m301036/LSAT/Data/Complete_TAVG_LatLong1.nc'
data_GISSTEMP = '/work/mh0033/m301036/LSAT/Data/air.2x2.250.mon.anom.land.nc'


tas_GISS  = xr.open_dataset(data_GISSTEMP)
tas_GISS  =tas_GISS.reindex(lat=tas_GISS['lat'][::-1])
tas_GISS
```

```python
tas_GISS_nh  = tas_GISS.sel(lat=slice(0,90))
tas_GISS_nh
```

```python
# Extract the desired time period
start_year = 1900
end_year = 2014
tas_GISS_nh = tas_GISS_nh.sel(time=slice(f'{start_year}-01-01', f'{end_year}-12-31'))
tas_GISS_nh
```

```python
# Calculate the monthly climatology
monthly_climatology = tas_GISS_nh['air'].groupby('time.month').mean(dim='time')
type(monthly_climatology)
print(monthly_climatology)

# Calculate the monthly anomalies by subtracting the climatology from the original data
monthly_anomalies = tas_GISS_nh['air'] - monthly_climatology
monthly_anomalies
# # Calculate the trend using linear regression
# time_values = pd.to_numeric(monthly_anomalies['time'].values)
# anomaly_values = monthly_anomalies.values
# trend = np.polyfit(time_values, anomaly_values, deg=1)
```
